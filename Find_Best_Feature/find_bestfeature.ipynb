```python
data_perf =[]
for file in os.listdir(basepath):
    data =pd.read_csv(basepath + file, sep='\t')
    data_perf.append(data)
    
```

```python
best_feature=[]
all_best_feature =[]
#for data in data_perf:
for j in range (len(data_perf)):
    best_feature = []
    for i in range (len(data_perf[j])):
         if data_perf[j]['Acc'].iloc[i] > 60 :
            best_feature.append(data_perf[j]['Id'].iloc[i])
    all_best_feature.append(best_feature)
```

```python

#for i in range (len(data_perf[0])):       # 35 loop
#    for j in (len(data_perf)):            # 15 loop
array_counter = pd.DataFrame(np.zeros((1, 35)))
for j in range (15):
    for i in range (35):
        if data_perf[j]['Id'].iloc[i] in all_best_feature[j]:
            array_counter[i] +=1 
        
        
```
