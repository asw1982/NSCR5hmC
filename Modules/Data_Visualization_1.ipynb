```python
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
from prettytable import PrettyTable
```

```python
def Get_STD(train_perf,test_perf,n_dat):
    train_perf_std=[]
    test_perf_std =[]
    for j in range (n_dat):  # 4 is number of data type
        type_perf_std1=[]
        type_perf_std2=[]
        for i in range (5):  # 5 is number of performance data in every fold validation
            type_perf_std1.append(statistics.stdev(train_perf[j][i]))
            type_perf_std2.append(statistics.stdev(test_perf[j][i]))
        train_perf_std.append(type_perf_std1)
        test_perf_std.append(type_perf_std2)
    return  train_perf_std, test_perf_std # its classified by type of performance 
```

```python
def Get_Mean(train_perf,test_perf, n_dat):
    train_perf_mean=[]
    test_perf_mean=[]
    for j in range (n_dat):  
        type_perf_mean1=[]
        type_perf_mean2=[]
        for i in range (5): 
            type_perf_mean1.append(statistics.mean(train_perf[j][i]))
            type_perf_mean2.append(statistics.mean(test_perf[j][i]))
        train_perf_mean.append(type_perf_mean1)
        test_perf_mean.append(type_perf_mean2)
    return  train_perf_mean, test_perf_mean # its classified by type of performance 
```

```python
# used for machine learning 
def Plot_Performances(train_perf,test_perf, model_title):
   
    train_perf_mean, test_perf_mean = Get_Mean(train_perf,test_perf)
    train_perf_std, test_perf_std = Get_STD(train_perf,test_perf)
    
    type_perf = ['ACC', 'SN', 'SP', 'MCC', 'AUC']
    x_indexes = np.arange(len(train_perf_mean[0])) #it should be 5 
    width = 0.2
    
    
    fig, axs = plt.subplots(1, 2,figsize=(12,7))
    fig.suptitle('Performances '+ model_title)
    #axs[0].figure(figsize=(7,5))
    #axs[1].figure(figsize=(7,5))
    
    axs[0].set_title('Train ' + model_title)
    axs[1].set_title('Test '+ model_title)
    
    
    axs[0].bar(x_indexes-3*width/2, train_perf_mean[0],yerr =train_perf_std[0],width=width,ecolor=['black'], color =['orange'], label = 'Bin1')
    axs[0].bar(x_indexes-width/2, train_perf_mean[1],yerr =train_perf_std[1],width=width,ecolor=['black'], color =['red'], label = 'Bin2')
    axs[0].bar(x_indexes+width/2, train_perf_mean[2],yerr =train_perf_std[2],width=width,ecolor=['black'], color =['blue'], label = 'NCP')
    axs[0].bar(x_indexes+3*width/2, train_perf_mean[3],yerr =train_perf_std[3],width=width,ecolor=['black'], color =['green'], label = 'EIIP')
    
    axs[0].legend(loc='lower right')
    axs[0].set(xlabel='perf types',ylabel ='perf in %')
    
    #axs[0].xticks(ticks =x_indexes, labels =type_perf)
    
    axs[1].bar(x_indexes-3*width/2, test_perf_mean[0],yerr =test_perf_std[0],width=width,ecolor=['black'], color =['orange'], label = 'Bin1')
    axs[1].bar(x_indexes-width/2, test_perf_mean[1],yerr =test_perf_std[1],width=width,ecolor=['black'], color =['red'], label = 'Bin2')
    axs[1].bar(x_indexes+width/2, test_perf_mean[2],yerr =test_perf_std[2],width=width,ecolor=['black'], color =['blue'], label = 'NCP')
    axs[1].bar(x_indexes+3*width/2, test_perf_mean[3],yerr =test_perf_std[3],width=width,ecolor=['black'], color =['green'], label = 'EIIP')
    
    axs[1].legend(loc='lower right')
    axs[1].set(xlabel='perf types',ylabel ='perf in %')
   
    plt.setp(axs, xticks=x_indexes, xticklabels=type_perf)
    
    plt.show()
       
```

```python
# used in deep learning
def Plot_Loss_Accuracy(k, array_history):
    
    for i in range(k):
        history_dict = array_history[i].history
        loss_values = history_dict['loss']
        val_loss_values = history_dict['val_loss']
        epochs = range(1, len(loss_values) + 1)
        plt.plot(epochs, loss_values, 'b', label='Training loss')
        plt.plot(epochs, val_loss_values, 'r', label='Validation loss')
        plt.title('Training and validation loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()
    
    for i in range(k):
    
        history_dict = array_history[i].history
        acc_values = history_dict['acc']
        val_acc_values = history_dict['val_acc']
        epochs = range(1, len(acc_values) + 1)
        plt.plot(epochs, acc_values, 'b', label='Training acc')
        plt.plot(epochs, val_acc_values, 'r', label='Validation acc')
        plt.title('Training and validation accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.show()
```

```python
def Create_Tables(train_perf,test_perf, model_title):
    train_perf_mean, test_perf_mean = Get_Mean(train_perf,test_perf)
    train_perf_std, test_perf_std = Get_STD(train_perf,test_perf)
    
    encoding_type =['BinaryI', 'BinaryII','NCP','EIIP']
    
    Table_train_m= PrettyTable(["mTrain Model Name", "Data Type", "ACC", "SN", "SP", "MCC", "AUC"])
    Table_test_m = PrettyTable(["mTest Model Name", "Data Type", "ACC", "SN", "SP", "MCC", "AUC"])
    Table_train_e = PrettyTable(["eTrain Model Name", "Data Type", "ACC", "SN", "SP", "MCC", "AUC"])
    Table_test_e = PrettyTable(["eTest Model Name", "Data Type", "ACC", "SN", "SP", "MCC", "AUC"])
    
    for i in range (len(encoding_type)):
        Table_train_m.add_row([model_title, encoding_type[i],str(round(train_perf_mean[i][0],3))
                         ,str(round(train_perf_mean[i][1],3)),str(round(train_perf_mean[i][2],3))
                         ,str(round(train_perf_mean[i][3],3)),str(round(train_perf_mean[i][4],3))])

        Table_test_m.add_row([model_title, encoding_type[i],str(round(test_perf_mean[i][0],3))
                         ,str(round(train_perf_mean[i][1],3)),str(round(test_perf_mean[i][2],3))
                         ,str(round(train_perf_mean[i][3],3)),str(round(test_perf_mean[i][4],3))])
        
        Table_train_e.add_row([model_title, encoding_type[i],str(round(test_perf_std[i][0],3))
                         ,str(round(train_perf_std[i][1],3)),str(round(test_perf_std[i][2],3))
                         ,str(round(train_perf_std[i][3],3)),str(round(test_perf_std[i][4],3))])
        
        Table_test_e.add_row([model_title, encoding_type[i],str(round(test_perf_std[i][0],3))
                         ,str(round(train_perf_std[i][1],3)),str(round(test_perf_std[i][2],3))
                         ,str(round(train_perf_std[i][3],3)),str(round(test_perf_std[i][4],3))])
        
    
    print(Table_train_m)
    print(Table_test_m)
    print(Table_train_e)
    print(Table_test_e)
```
