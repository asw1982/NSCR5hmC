```python
def Machine_Learning(k,ML_model,t_vec_train,t_vec_val,t_vec_test,type_ML,file_dir):
    total_y_pred_train =[]
    total_y_pred_test =[]
    total_y_true_train =[]
    total_y_true_test =[]
    
    for i in range(k):
        model = ML_model

        # data have 5 different arrangements (5 fold)
        # Train the model using the training sets
        features = t_vec_train[i][0]
        label = t_vec_train[i][1]
        model.fit(features,label)

        #Predict Output 
        # evaluation of training session
        features_v = t_vec_val[i][0]
        label_v  = t_vec_val[i][1]
        
        features_train = np.concatenate((features,features_v),axis=0)
        
        y_pred_train= model.predict(features_train)
        y_true_train =np.concatenate((label,label_v),axis=0) 
        
               
        #evaluation of test session
        
        y_pred_test= model.predict(t_vec_test[0])
        y_true_test = t_vec_test[1]
        
        total_y_pred_train.append(y_pred_train)
        total_y_pred_test.append(y_pred_test)
        total_y_true_train.append(y_true_train)
        total_y_true_test.append(y_true_test)
        # save the model 
        if 'n' in type_ML:    
            filename = file_dir +'n_model/'+ type_ML + '_fold'+ str(i)+'.joblib'
        else :
            filename = file_dir +'_model/'+ type_ML + '_fold'+ str(i)+'.joblib'
        Save_Model(model, filename)
    
    return total_y_pred_train,total_y_true_train,total_y_pred_test,total_y_true_test
```

```python
def KNN(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model =KNeighborsClassifier(n_neighbors=21)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)

def RF(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model =RandomForestClassifier(max_depth=10, random_state=0)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)

def SGD(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model =SGDClassifier(max_iter=1000, tol=1e-3)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)

def GB(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model =GradientBoostingClassifier(n_estimators=100, learning_rate=0.01
                                                 ,max_depth=10, random_state=0)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)

def AB(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model =AdaBoostClassifier(n_estimators=100, random_state=0)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)

def SVM_C(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model =svm.SVC(kernel='linear')
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)

def LR(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model =LogisticRegression(random_state=0)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)

def XGB(k,t_vec_train,t_vec_val,t_vec_test):
    ML_model = XGBClassifier(objective="binary:logistic", random_state=0)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)  
```

```python
def Run_ML(k,t_vec_train,t_vec_val,t_vec_test,type_ML):
    if 'KNN' in type_ML :
        ML_model =KNeighborsClassifier(n_neighbors=21)
        file_dir ='KNN'
    if 'RF' in type_ML :
        ML_model =RandomForestClassifier(max_depth=10, random_state=0)
        file_dir ='RF'
    if 'SGD' in type_ML :
        ML_model =SGDClassifier(max_iter=1000, tol=1e-3)
        file_dir ='SGD'
    if 'GB' in type_ML :
        ML_model =GradientBoostingClassifier(n_estimators=100, learning_rate=0.01
                                                 ,max_depth=10, random_state=0)
        file_dir ='GB'
    if 'AB' in type_ML :
        ML_model =AdaBoostClassifier(n_estimators=100, random_state=0)
        file_dir ='AB'
    if 'SVM' in type_ML :
        ML_model =svm.SVC(kernel='linear')
        file_dir ='SVM'
    if 'XGB' in type_ML :
        ML_model =XGBClassifier(objective="binary:logistic", random_state=0)
        file_dir ='XGB'
    if 'Stacked' in type_ML :
    #define the base models
        level0 = list()
        level0.append(('gb', GradientBoostingClassifier(n_estimators=100, learning_rate=0.01
                                                 ,max_depth=10, random_state=0)))
        level0.append(('ab', AdaBoostClassifier(n_estimators=100, random_state=0)))
        level0.append(('xgb',XGBClassifier(objective="binary:logistic", random_state=0)))
        level0.append(('rf', RandomForestClassifier(max_depth=10, random_state=0)))
   
    # define meta learner model
        level1 = LogisticRegression(C=0.1)
    # define the stacking ensemble
        ML_model = StackingClassifier(estimators=level0, final_estimator=level1)   
        file_dir ='Stacked'
    return Machine_Learning(k,ML_model,t_vec_train,t_vec_val,t_vec_test,type_ML,file_dir)  
```

```python
def Stacking(k,t_vec_train,t_vec_val,t_vec_test):
   # define the base models
    level0 = list()
    level0.append(('gb', GradientBoostingClassifier(n_estimators=100, learning_rate=0.01
                                                 ,max_depth=10, random_state=0)))
    level0.append(('ab', AdaBoostClassifier(n_estimators=100, random_state=0)))
    level0.append(('xgb',XGBClassifier(objective="binary:logistic", random_state=0)))
    level0.append(('rf', RandomForestClassifier(max_depth=10, random_state=0)))
   
    # define meta learner model
    level1 = LogisticRegression()
    # define the stacking ensemble
    ML_model = StackingClassifier(estimators=level0, final_estimator=level1)
    return Machine_Learning_0(k,ML_model,t_vec_train,t_vec_val,t_vec_test)  
```
