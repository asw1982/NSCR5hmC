```python
def plot_ROC(total_data_roc, method):
    i =0
    for data_roc in total_data_roc:
    # plot ROC 
        fpr,tpr = data_roc[0], data_roc[1]
        plt.plot(fpr,tpr,marker='.',linestyle='--', label=method+'_fold_'+ str(i))
        i = i+1
# axis labels
    plt.plot([0,1],[0,1],marker='.',linestyle='--',color='k')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
# show the legend
    plt.legend()
# show the plot
    plt.show()
```

```python
list_method = ['AB', 'GB','RF','XGB']
AB_auc = AB_total_performances_test3[4]
GB_auc = GB_total_performances_test3[4]
RF_auc = RF_total_performances_test3[4]
XGB_auc = XGB_total_performances_test3[4]
total_data_AUC = [AB_auc, GB_auc, RF_auc, XGB_auc]
total_data_ROC =[AB_total_ROC_test3,GB_total_ROC_test3,RF_total_ROC_test3,XGB_total_ROC_test3]
def plot_ROC_1(list_method, fold):
    i =0
    for method in list_method:
    # plot ROC 
        data_roc=total_data_ROC[i]
        data_auc=total_data_AUC[i]
        fpr,tpr =data_roc[fold] 
        plt.plot(fpr,tpr,marker='.',linestyle='--', label=method+'_fold_'+ str(fold) +" AUC="+str(data_auc[fold]))
        i=i+1
        
# axis labels
    plt.plot([0,1],[0,1],marker='.',linestyle='--',color='k')
    plt.title('ROC Curve fold ' + str(fold))
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
# show the legend
    plt.legend()
# show the plot
    plt.show()
```

```python
for i in range(10):
    plot_ROC_1(list_method, fold=i)
```
